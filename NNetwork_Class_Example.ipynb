```python
import pandas as pd
import numpy as np
```

```python
from sklearn import preprocessing
```

```python
from keras.layers import Dense
```

```python
from keras.models import Sequential
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#from scikeras.wrappers import KerasClassifier
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
seed = 7
```

```python
dataset = pd.read_csv("pima-indians-diabetes.csv", header=None)
```

```python
dataset
```

```python
scaler = preprocessing.MinMaxScaler()
```

```python
scaled_dataset = scaler.fit_transform(dataset)
```

```python
X = scaled_dataset[:, :8]
```

```python
y = scaled_dataset[:, 8]
```

```python
model = Sequential()
```

```python
model.add(Dense(12, input_dim = 8, kernel_initializer = 'uniform', 
         activation = 'sigmoid'))
```

```python
model.add(Dense(1, kernel_initializer = 'uniform', 
         activation = 'sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs = 50, batch_size=10, verbose=2)
```

```python
def create_model(neurons = 1):
    model = Sequential()
    model.add(Dense(neurons, input_dim = 8, kernel_initializer = 'uniform', 
         activation = 'sigmoid'))
    model.add(Dense(1, kernel_initializer = 'uniform', 
         activation = 'sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn = create_model, epochs = 100, batch_size = 10, verbose =2)
```

```python
neurons = [1, 5, 10, 15, 20, 25, 30]
```

```python
param_grid = dict(neurons=neurons)
```

```python
grid = GridSearchCV(estimator=model, param_grid = param_grid)
```

```python
grid_result = grid.fit(X, y)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params))
```

```python
grid_result.cv_results_
```

```python
means = grid_result.cv_results_["mean_test_score"]
stds = grid_result.cv_results_["std_test_score"]
params = grid_result.cv_results_["params"]
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with %r" % (mean, stdev, param))
```
